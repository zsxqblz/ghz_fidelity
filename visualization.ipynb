```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = "data/200207_nf"

meas_fidelity_raw = np.loadtxt(file_name+"_scanx.csv",skiprows=1,delimiter=",")
meas_fidelity_l = meas_fidelity_raw
nsites_raw = np.loadtxt(file_name+"_scany.csv",skiprows=1,delimiter=",")
nsites_l = nsites_raw

meas_fidelity_length = len(meas_fidelity_l)
nsites_length = len(nsites_l)

data_raw =  np.loadtxt(file_name+"_data.csv",skiprows=1,delimiter=",")
population_ave_l = data_raw[:,0].reshape((meas_fidelity_length,nsites_length))
population_std_l = data_raw[:,1].reshape((meas_fidelity_length,nsites_length))
coherence_ave_l = data_raw[:,2].reshape((meas_fidelity_length,nsites_length))
coherence_std_l = data_raw[:,3].reshape((meas_fidelity_length,nsites_length))
```

```python
for i in range(3,20,1):
    plt.plot(meas_fidelity_l,population_ave_l[:,i-1],'b')

plt.rcParams["figure.figsize"] = (5,4)
plt.rcParams.update({'font.size': 22})
plt.xlim([0.9,1])
plt.ylim([0.1,1])
plt.yscale('log')
plt.yticks([0.1,0.2,0.3,0.4,0.6,0.8,1],["0.1","0.2","0.3","0.4","0.6","0.8","1"])
plt.xlabel("measurement fidelity")
plt.ylabel("population")
```

```python
for i in range(3,20,1):
    plt.plot(meas_fidelity_l,coherence_ave_l[:,i-1])
    
plt.xlim([0.9,1])
```
